# Merging Actor and Description Data

```python
#imports
import pandas as pd
import numpy as np
```

```python
#reading in actor df
actor = pd.read_csv('data/df_with_actor_data.csv',index_col=0)
print(actor.shape)
actor.head()
```

```python
#reading in description data df
description = pd.read_csv('data/df_with_movie_desc.csv')
print(description.shape)
description.head()
```

```python
#creating id column to merge dataframes on
desc_dict = {}
for index, row in description.iterrows():
    desc_dict[row['id']] = row['movie_desc']
```

```python
#adding description data to actor df using id
actor['desc'] = actor['id'].map(desc_dict)
```

```python
# function that gets movie id from cast_link
def imdb_movie_id(link):
    if 'https://www.imdb.com/title/' in link:
        return link.split('/')[4]
    else:
        return 'no ID'
```

```python
#rewriting movie id with proper function
actor['imdb_movie_id'] = actor['imdb_cast_link'].apply(imdb_movie_id)
```

```python
#renaming combined data df
df = actor
```

```python
#saving data to csv
df.to_csv('data/total_webscraped.csv')
```
